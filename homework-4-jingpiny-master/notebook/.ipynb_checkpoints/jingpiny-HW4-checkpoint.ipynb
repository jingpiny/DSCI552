1. The LASSO and Boosting for Regression
(a) Use the first 1495 rows of data as the training set and the rest as the test set.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# get features'name from .names
name = list()
with open("../data/communities.names") as file:
    for row in file:
        if row.startswith("@attribute"):
            row_list = row.split(" ")
            name.append(row_list[1])
print("There are", str(len(name)), "predictors.")
```

```python
# get the data from .data
data = pd.read_csv("../data/communities.data", names=name)
data = data.replace('?', np.nan)
data
```

```python
# split the training and test set
train = data.iloc[:1495]
test = data.iloc[1495:]
print("There are", str(len(train)), "points for training, and", str(len(test)), "points for test set.")
```

(b) The data set has missing values. Use a data imputation technique to deal with the missing values in the data set. The data description mentions some features are nonpredictive. Ignore those features.

```python
# remove the first five features
feature_remove = name[:5]
train = train.drop(feature_remove, axis=1)
test = test.drop(feature_remove, axis=1)
```

```python
from sklearn.impute import SimpleImputer

imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
imp_mean.fit(train)
train = pd.DataFrame(imp_mean.transform(train)).round(2)
train.columns = name[5:]
test = pd.DataFrame(imp_mean.transform(test)).round(2)
test.columns = name[5:]
print(train)
print(test)
```

```python
# split the goal (Y) with the features (X)
trainY=train['ViolentCrimesPerPop']
testY=test['ViolentCrimesPerPop']
testX=test.drop('ViolentCrimesPerPop',axis=1)
trainX=train.drop('ViolentCrimesPerPop',axis=1)
```

(c) Plot a correlation matrix for the features in the data set.

```python
f = plt.figure(figsize=(30, 30))
plt.matshow(trainX.corr(), fignum=f.number)
plt.xticks(range(trainX.select_dtypes(['number']).shape[1]), trainX.select_dtypes(['number']).columns, fontsize=14,rotation=90)
plt.yticks(range(trainX.select_dtypes(['number']).shape[1]), trainX.select_dtypes(['number']).columns, fontsize=14)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Correlation Matrix', fontsize=16);
```

(f) Fit a linear model using least squares to the training set and report the test error.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model = LinearRegression().fit(trainX, trainY)
predY = model.predict(testX)
test_error=mean_squared_error(testY, predY)
print("The test error of least squares model is: ", test_error)
```

(g) Fit a ridge regression model on the training set, with λ chosen by cross-validation. Report the test error obtained.

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import RidgeCV

#ridge cross-validation
cv = RidgeCV(alphas=(0.1,1.0,10.0), cv=5).fit(trainX, trainY)
best_alpha = cv.alpha_
print("The best λ is: ", best_alpha)

ridge = Ridge(alpha = best_alpha).fit(trainX, trainY)
ridge_pred =ridge.predict(testX)
ridge_test_error = mean_squared_error(testY,ridge_pred)
print("The test error of least squares model is: ", ridge_test_error)
```

(h) Fit a LASSO model on the training set, with λ chosen by cross-validation. Report the test error obtained, along with a list of the variables selected by the model. Repeat with standardized features. Report the test error for both cases and compare them.

```python
from sklearn.linear_model import Lasso
from sklearn.linear_model import LassoCV

# Lasso cross_validation
lassocv = LassoCV(alphas=(0.001,1,10.0), cv=5).fit(trainX,trainY)
lasso_alpha = lassocv.alpha_
print("The best λ is: ", lasso_alpha)

lasso = Lasso(alpha = lassocv.alpha_).fit(trainX, trainY)
lasso_pred =lasso.predict(testX)
lasso_test_error = mean_squared_error(testY,lasso_pred)
print("The test error of Lasso model is: ", lasso_test_error)
```

```python
name = name[5:]
```

```python
index = list()
count = 0
for coef in lasso.coef_:
    if coef != 0:
        index.append(count)
    count += 1

selected_features = list()
for i in index:
    selected_features.append(name[i])
print(selected_features)
```

```python
# standardized Lasso
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(trainX)
trainX_std = scaler.transform(trainX)
testX_std = scaler.transform(testX)

lassocv1 = LassoCV(alphas=(0.001,1,10.0), cv=5).fit(trainX_std, trainY)
lassocv1_alpha = lassocv1.alpha_
print("The best λ is:", lassocv1_alpha)

lasso1 = Lasso(alpha = lassocv1_alpha).fit(trainX_std, trainY)
lasso1_pred =lasso1.predict(testX_std)
lasso1_test_error =  mean_squared_error(testY, lasso1_pred)
print("The test error for lasso model of standardized features is ", lasso1_test_error)
```

```python
index = list()
count = 0
for coef in lasso1.coef_:
    if coef != 0:
        index.append(count)
    count += 1

selected_features = list()
for i in index:
    selected_features.append(name[i])
print(selected_features)
```

<b>Comparing models using the normal features and the standardized features, we can see that there are more features selected after the standardization. The test error for the model also drops.
